```python
def conll_read_sentence(file_handle):
	sent = []
	for line in file_handle:
		line = line.strip('\n')
		if line.startswith('#') is False :
			toks = line.split("\t")				
			if len(toks) == 1:
				return sent 
			else:
				if toks[0].isdigit() == True:
					sent.append(toks)
	return None
```

```python
data = []
with open('../UD_data/UD_English-GUMReddit/en_gumreddit-ud-train.conllu') as f:
  for line in f:
    sent = conll_read_sentence(f)
    print(sent)
    if sent is not None:
      for tok in sent:
        data.append(tok[1])

print(data)
print(len(set(data)))
```
